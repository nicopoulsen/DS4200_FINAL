## Visualizations for Group Project Proposal
### Predicting Car Prices: The Role of Features and Market Trends


### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('car_price_dataset.csv')
df
```

## Visualization 1 (Potential)

```python
plt.figure(figsize=(10,6))

plt.scatter(df['Mileage'], df['Price'], alpha=0.2)

sns.kdeplot(x=df['Mileage'], y=df['Price'], cmap="Reds", fill=True, alpha=0.5)

plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Scatter Plot of Mileage vs Price with KDE')
plt.show()
```

## Visualization 2 (Potential)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Transmission'], y=df['Price'])
plt.title('Impact of Transmission Type on Price')
plt.xlabel('Transmission Type')
plt.ylabel('Price')
plt.show()
```

## Visualization 3 (Potential)

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['Year'], y=df['Price'])
plt.title('Price Depreciation Over Years')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price')
plt.show()
```

## Visualization 4 (Potential) 
We have noticed that the distribution is very similar so the visualization looks brand alone isn't a strong differentiator of car prices in our dataset Reconsider dataset or visualization?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

sns.boxplot(x=df['Brand'], y=df['Price'])

plt.xticks(rotation=45)

plt.xlabel('Car Brand')
plt.ylabel('Price')
plt.title('Car Price Distribution by Brand')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Price', hue='Brand')
plt.title('Brand Price Depreciation Over Time')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
